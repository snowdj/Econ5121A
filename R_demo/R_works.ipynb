Get Federal Reserve data [http://research.stlouisfed.org/fred2/](http://research.stlouisfed.org/fred2/). For example, the following code in one line download the 10-Year Treasury Constant Maturity Rate (DGS10) and Consumer Price Index for All Urban Consumers: All Items (CPIAUCSL).

```r
library(quantmod)
getSymbols(Symbols = c("CPIAUCSL", "DGS10"), env = .GlobalEnv, src="FRED" ) 
# the line above will generate to objects "CPIAUCSL" and "DGS10"
plot(DGS10)
```
